```python
lgbm = create_model('lightgbm')
```

```python
tuned_lgbm = tune_model(lgbm)
```

```python
add_metric('logloss', 'Log Loss', log_loss, greater_is_better=False, target='pred_proba')
```

```python
top3 = compare_models(n_select = 3)
tuned_top3 = [tune_model(i, optimize='Log Loss') for i in top3]
blender = blend_models(tuned_top3)
stacker = stack_models(tuned_top3)
best_logloss_model = automl(optimize = 'logloss')
```

```python
final_model = finalize_model(blender)
prediction = predict_model(final_model, data=test)
```

```python
prep_pipe = get_config("prep_pipe")
prep_pipe.steps.append(['trained_model', final_model])
prections = prep_pipe.predict_proba(test)
prections
```

```python
i, j, k = [], [], []
for row in prections:
  i.append(row[0])
  j.append(row[1])
  k.append(row[2])

len(i), len(j), len(k)
```

```python
submit['0'] = i
submit['1'] = j
submit['2'] = k
submit.head(10)
```
